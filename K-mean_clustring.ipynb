```
#k-means clustring example home rental
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
#sample
df=pd.read_csv('/content/ice-cream.csv')
print(df.head())
```

```
#create k-mean model
kmean = KMeans(n_clusters=3,random_state=42)
df['cluster']= kmean.fit_predict(df[['Temperature','IceCreamsSold']])
```

```
#centroids
centroids =kmean.cluster_centers_
```

```

#plot clusters
plt.scatter(df['Temperature'],df['IceCreamsSold'],c=df['cluster'],cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,1],color='black',marker='x',s=100)#centroids
plt.xlabel('Temperature')
plt.ylabel('IceCreamsSold')
plt.title("K-Mean clustring example")
plt.show()
```
