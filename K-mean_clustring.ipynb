```
#k-means clustring example group customers
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```
#sample
df=pd.read_csv('/content/Mall_Customers.csv')
print(df.head())
```

```
#create k-mean model
kmean = KMeans(n_clusters=3,random_state=42)
df['cluster']= kmean.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
```

```
#centroids
centroids =kmean.cluster_centers_
```

```

#plot clusters
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'],c=df['cluster'],cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,1],color='black',marker='x',s=100)#centroids
plt.xlabel('Annual Income (k$)e')
plt.ylabel('Spending Score (1-100)')
plt.title("K-Mean clustring example")
plt.show()
```
